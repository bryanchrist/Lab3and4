I do not see any countries that are present in both unmatched data sets. For countries in World Bank but not in V-Dem, there are three clear patterns in V-Dem's data collection that led to these gaps. The first is that V-Dem didn't collect data from former individual Soviet States, while the World Bank did. The second is that V-Dem did not collect data from very small countries like Monaco, while the World Bank did. Finally, V-Dem also did not collect data from countries that came into existence during part of this timeframe (1960-2021) like South Sudan, while the World Bank did. For countries in V-Dem that are not in the World Bank data, these countries are technically not independent (ex. Taiwan and Hong Kong are technically part of China). It is likely the World Bank does not collect data from these countries individually because they are not recognized as independent, while V-Dem does collect data from them.

## Question 3
Initially, the data are not in third normal form. The data are in first normal form, as each table has a primary key, the values inside every cell are atomic, and there are no repeating groups. However, the data are not in second normal form, as in the timeseries table, country_name_wb and country_name_vdem are dependent on part of the primary key (just country_code). Dropping these two country variables puts all three tables in second normal form. 

After dropping these variables, the tables are also in third normal form because there are no functional dependencies outside of the primary key. For example, in the timeseries data, there are no columns that depend on anything else other than year and country_code.

## Number 6 C

```python
myquery = '''
SELECT 
   c.region, 
   SUM(t.co2_emissions) AS co2_emissions
FROM timeseries t
INNER JOIN country c
    ON t.country_code = c.country_code
WHERE year = '2019'
GROUP BY region
ORDER BY co2_emissions DESC
'''

pd.read_sql(myquery, con=engine)
```

## Number 6 D

```python
myquery = '''
WITH past AS (SELECT 
    t.country_code, 
    t.democracy
FROM timeseries t
WHERE year = '1960')

SELECT 
    c.country_name_wb as country,
    past.democracy as past_democracy,
    pr.democracy as present_democracy,
    SUM(pr.democracy - past.democracy) AS difference
FROM past
JOIN (SELECT 
    t.country_code, 
    t.democracy
FROM timeseries t
WHERE year = '2021') pr
    ON past.country_code = pr.country_code
JOIN country c
    ON c.country_code=past.country_code
GROUP BY past.country_code, past.democracy, pr.democracy, c.country_name_wb
ORDER BY difference DESC
LIMIT 10
'''
pd.read_sql(myquery, con=engine)
```

## Question 6 E

By count of countries, what is the most commonly used currency in the world?

```python
myquery = '''
SELECT  
   c.currency_unit,
   COUNT(c.currency_unit) AS number_of_countries_used
FROM timeseries t
INNER JOIN country c
    ON t.country_code = c.country_code
GROUP BY c.currency_unit
ORDER BY number_of_countries_used DESC
LIMIT 1
'''

pd.read_sql(myquery, con=engine)
```

## Question 6 F
The GINI index measures the amount of economic inequality in a country. The higher the index, the greater the economic disparity between rich and poor. How does the average GINI index compare across income groups (see the country table) in 2019?

```python
myquery = '''
SELECT
   c.income_group,
   AVG(t.gini_index) as avg_gini
FROM timeseries t
INNER JOIN country c
    ON t.country_code = c.country_code
GROUP BY c.income_group
ORDER BY avg_gini DESC
'''

pd.read_sql(myquery, con=engine)
```

Extra credit: Some countries include the word 'Republic' or 'Democratic' in their official names. Other countries do not. Which of these two groups of countries had a higher average level of democratic quality in 2021? (It's OK if you use two queries to provide this answer.) [5 bonus points]

Answer: Based on these two queries, the countries with the word Republic had a higher average level of democratic quality in 2021. 

```python
myquery = '''
SELECT
   AVG(t.democracy) as average_democracy_score
FROM timeseries t
INNER JOIN country c
    ON t.country_code = c.country_code
WHERE c.country_name_wb LIKE '%%Rep%%' AND year = '2021'
'''
pd.read_sql(myquery, con=engine)
```
